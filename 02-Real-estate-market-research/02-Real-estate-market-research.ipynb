# Исследование объявлений о продаже квартир

**Цель исследования:**

Установить параметры для определения рыночной стоимости недвижимости. Результаты исследования будут учтены при построении автоматизированной системы для определения рыночной стоимости: она отследит аномалии и мошенническую деятельность.

**Исходные данные:** 

Данные сервиса Яндекс.Недвижимость — архив объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктах за несколько лет. По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые получены автоматически на основе картографических данных.

**Задачи:** 

- Изучение параметров объектов недвижимости.
- Изучение время продажи квартиры.
- Выявление редких и выбивающихся значений.
- Изучение зависимости различных параметров квартир и объявлений от их рыночной стоимости.
- Определение средней цены квадратного метра.
- Изучение предложения о продаже квартир в центре Санкт-Петербурга.

## Обзор данных

### Описание данных
​
**Описание исходной таблицы с данными**

- airports_nearest — расстояние до ближайшего аэропорта в метрах (м)
- balcony — число балконов
- ceiling_height — высота потолков (м)
- cityCenters_nearest — расстояние до центра города (м)
- days_exposition — сколько дней было размещено объявление (от публикации до снятия)
- first_day_exposition — дата публикации
- floor — этаж
- floors_total — всего этажей в доме
- is_apartment — апартаменты (булев тип)
- kitchen_area — площадь кухни в квадратных метрах (м²)
- last_price — цена на момент снятия с публикации
- living_area — жилая площадь в квадратных метрах(м²)
- locality_name — название населённого пункта
- open_plan — свободная планировка (булев тип)
- parks_around3000 — число парков в радиусе 3 км
- parks_nearest — расстояние до ближайшего парка (м)
- ponds_around3000 — число водоёмов в радиусе 3 км
- ponds_nearest — расстояние до ближайшего водоёма (м)
- rooms — число комнат
- studio — квартира-студия (булев тип)
- total_area — площадь квартиры в квадратных метрах (м²)
- total_images — число фотографий квартиры в объявлении

```python
def display_datainfo(path, sep):
    """функция для вывода датасета, описания, подсчета значений в столбцах
    
    принимает на вход путь к файлу path и разделитель sep"""
    
    try:
        df = pd.read_csv(path, sep)
    except:
        df = pd.read_csv('/datasets/'+path, sep)
    print('\033[1m'+'Первые 5 строк данных'+'\033[0m')
    display(df.head())
    print('_'*50)
    print('\033[1m'+"Типы и размер данных"+'\033[0m')
    display(df.info())
    print('_'*50)
    print('\033[1m'+"Статистика"+'\033[0m')
    display(df.describe())
    print('_'*50)
    print('\033[1m'+ 'Всего пропусков:'+'\033[0m')
    display(df.isna().sum())
    print('_'*50)
    print('\033[1m'+f'Полные дубликаты:{df.duplicated().sum()}'+'\033[0m')
    print('_'*50)
    print('\033[1m'+'Значения столбцов:'+'\033[0m')
    print(' ')
    for col in df.columns:
        print('\033[1m'+str(col)+'\033[0m')
        print(df[col].value_counts())
        print('_'*50)
        print('')
    return df
```

49 Пропусков в названиях населенных пунктов могли появится как в результате ошибок ввода, так и выгрузки данных. В значениях 'airports_nearest' встречаются повторы значений с точностью до 1 метра. Заполним пропуски значением 'Неизвестен'.

```python
data['locality_name'] = data['locality_name'].fillna('Неизвестен')
```

По повторяющимся значениям вычислим пропущенные названия населенных пунктов.

```python
#Используем функцию для автоматического поиска и замены значений 'Неизвестен' на моды по выборке по расстоянию до аэропорта.
#Опытным путем установили, что результат замены модальными значениями оптимальнее, чем медианами.
def locality_fillin(row):
    airports_nearest = row['airports_nearest']
    locality_name = row['locality_name']
    for row in data:
        try:
            if locality_name == 'Неизвестен':
                locality_name = statistics.mode(data[(data['airports_nearest'] == airports_nearest)
                                                    & (data['locality_name'] != 'Неизвестен')
                                                    ]['locality_name'])  
                return locality_name
            return locality_name
        except:
            return 'Неизвестен'
    
data['locality_name'] = data.apply(locality_fillin, axis = 1)
```

Нулевые (пропущенные) значения заменим на медианы по выборке данных по наиболее влияющим параметрам: общая площадь и число комнат. Новые данные будут справедливы также для студий и квартир со свободной планировкой.

```python
#Используем функцию для автоматического поиска и замены нулевых значений. Выборку сделаем с помощью фильтра по трем параметрам.
def living_area_fillin(row):
    living_area = row['living_area']
    total_area = row['total_area']
    rooms = row['rooms']
    for row in data:
        if living_area == 0:
            living_area = data[(data['living_area'] != 0) 
                               & ((total_area-5) <= data['total_area'])
                               & (data['total_area'] <= (total_area+5))
                               & (data['rooms'] == rooms)]['living_area'].median()         
            return living_area
        return living_area
     
data['living_area'] = data.apply(living_area_fillin, axis = 1)
```

```python
#Используем функцию для автоматического поиска и замены нулевых значений.
#Для расчета площади кухни найдем "kitchen_share" - среднее значение доли кухни в нежилой площади квартиры.
kitchen_in = data[data['kitchen_area']!=0]
kitchen_share = kitchen_in['kitchen_area'].sum()/(kitchen_in['total_area'].sum()-kitchen_in['living_area'].sum())
print(kitchen_share)

def kitchen_area_fillin(row):
    kitchen_area = row['kitchen_area']
    total_area = row['total_area']
    living_area = row['living_area']
    for row in data:
        if kitchen_area == 0:
            kitchen_area = (total_area-living_area)*kitchen_share  
            return kitchen_area
        return kitchen_area     
     
data['kitchen_area'] = data.apply(kitchen_area_fillin, axis = 1)
```

- В других столбцах наличие NaN значений может говорить о невозможности определить объекты в радиусе 3 км по причине отсутствия таковых, либо **из-за отсутствия или неверно указанного адреса** объекта недвижимости.Кроме того пропуски легко объяснить **несовершенством системы автоматизированного сбора данных** локализации, а так же неполнотой данных о таких объектах на картах. Так же возможно **система настроена** таким образом, что расстояния, например больше 100 км, **не отображаются**.
- **Незначительно  выражена зависимость числа балконов от этажности здания**. **Пропуски** в данных о числе балконов носят **случайный характер**, но чаще встречаются в пятиэтажках. Заполнили пропуски нулями. Изменили тип данных на целочисленный.
- В столбце **"апартаменты"** заменили пропуски нулями и изменили тип данных на булевый.
- Пропуски в столбце **'число этажей'** заполнили медианным значением по выборке для каждого этажа. Изменили тип данных на целочисленный int.
- Пропуски в **названиях населенных пунктов** могли появится как в результате ошибок ввода, так и выгрузки данных. Половину пропусков заполнили модами по сгруппированным по расстоянию до аэропорта данным. Остальные заполнили значением 'Неизвестен'.
- Нули в столбце **"число комнат"** заменили на 1, т.к. значения находится в данных о квартирах-студиях и со свободной планировкой.
- Пропуски в данных о **жилой площади** заменили на медианы по выборке данных по наиболее влияющим параметрам: общая площадь и число комнат.
- Пропуски в данных о **площади кухни**, где это возможно, заполнили как среднее отношение площади кухни к нежилой площади.
- В столбце **"дата публикации"** изменили тип данных на datetime.
- Вероятнее всего пропуски данных **"число дней публикации"** являются результатом ошибок при выгрузке данных либо относятся к объявлениям, которые не сняты с публикации на момент выгрузки. Необходимо наладить процесс выгрузки данных таким образом, чтобы избежать появления подобных проблем. Мы изменили тип данных на целочисленный.
- Тип данных в столбце **окончательная цена** так же изменили на целочисленный.

## Рассчёты и добавление результатов в таблицу

### Цена квадратного метра

Расчитаем цену квадратного метра для каждой квартиры как отношение итоговой цены к общей площади и добавим новый столбец 'price_per_sqm' с данными в рублях в таблицу.

```python
data['price_per_sqm'] = data['last_price']/data['total_area']
```

```python
#Для замены типа данных на int используем метод .astype('int'). 
data['price_per_sqm'] = data['price_per_sqm'].astype('int')
```

### День недели, месяц и год публикации

Вычислим день недели публикации объявления методом .dt.weekday и добавим новый столбец 'weekday_exposition' в таблицу. Численные обозначения дней недели заменим на строковые с помощью функции "rename_weekday" для столбца 'weekday_exposition'.

```python
data['weekday_exposition'] = data['first_day_exposition'].dt.weekday
```

```python
def rename_weekday(weekday):
        if weekday == 0:
            day_name = 'понедельник'
        elif weekday == 1: 
            day_name = 'вторник'
        elif weekday == 2: 
            day_name = 'среда'
        elif weekday == 3: 
            day_name = 'четверг'    
        elif weekday == 4: 
            day_name = 'пятница'
        elif weekday == 5: 
            day_name = 'суббота'
        elif weekday == 6: 
            day_name = 'воскресенье'
        else:
            day_name = 'ошибка'
        return day_name
```

Вычислим месяц и год публикации объявления аналогичным образом методам .dt.month и .dt.year и добавим новые столбцы 'month_exposition' и 'year_exposition' в таблицу.

```python
data['weekday_exposition'] = data['weekday_exposition'].apply(rename_weekday)
data['month_exposition'] = data['first_day_exposition'].dt.month
data['year_exposition'] = data['first_day_exposition'].dt.year
```

### Этаж квартиры

Проведем категоризацию данных об этажах квартир по трём категориям: первый, последний, другой.

```python
#Функция для категоризации данных об этажах.
def categorize_floor(row):
    floor = row['floor']
    floors_total = row['floors_total']
    for row in data:
        if floor == 1:
            floor_type = 'первый'
        elif floor == floors_total: 
            floor_type = 'последний'     
        else:
            floor_type = 'другой' #Проверочное значение
        return floor_type 
```

```python
#Применим функцию для стобца'floor_category'     
data['floor_category'] = data.apply(categorize_floor, axis = 1)

#Проверим результат применения функции
data['floor_category'].value_counts()
```

### Отношение жилой площади и площади кухни к общей площади

Вычислим почередно отношение жилой площади и площади кухни к общей площади, результаты округлим до двух знаков после точки и поместим данные во вновь созданные столбцы 'living_area_share' и 'kitchen_area_share'.

```python
data['living_area_share'] = round(data['living_area']/data['total_area'], 2)
data['kitchen_area_share'] = round(data['kitchen_area']/data['total_area'], 2)

#Выведем на экран дополненную таблицу данных.
data.head()
```

- Провели категоризацию этажей квартиры. Добавили в таблицу категори этажей: первый, последний, другой.

Коэффициенты корреляции для цены квартир на момент снятия объявления посчитаны только для численных величин. Для анализа всех параметров поочередно выведем на экран диаграммы рассеяния для всех столбцов методом .plot. Используем функцию.

```python
#Функция для создания диаграммы рассеяния
def price_scatter(column):
    return price_affecting_data.plot(y='last_price', 
                                    x=column, 
                                    kind='scatter', 
                                    title = 'Корреляция last_price с '+column, 
                                    grid=True,
                                    figsize=(15,4.5),
                                    alpha=0.25);
```

Для оценки влияния различных параметров на цену квартиры сгруппируем такие параметры по признакам методом .groupby. По каждой группе в таблице посчитаем минимальную и максимальную цены, среднюю цену и медиану, а так же удельный вес (долю) каждой группы во всей совокупности групп. Используем функцию.

```python
#Функция groupby с установленными названиями столбцов сортирует таблицу по столбцу "category_ratio".
def price_summary(column):
    summary=price_affecting_data.groupby(column).agg(
            category_ratio=("last_price", "count"),
            min_price=("last_price", "min"),
            max_price=pd.NamedAgg(column="last_price", aggfunc="max"),
            mean_price=("last_price", np.mean),
            median_price=pd.NamedAgg(column="last_price", aggfunc='median')
    )
    summary['mean_price']= round(summary['mean_price'])
    summary['category_ratio']=round(summary['category_ratio']/summary['category_ratio'].sum()*100,1)
    summary['median_to_mean']= round(summary['median_price']/summary['mean_price'],2)
    print('Зависимость цен на недвижимость от',column)
    return summary.sort_values(by = 'category_ratio', ascending = False)
```

* Средняя цена квадратного метра - 99,4 тыс.руб. медиана - 95 тыс.руб. Цена до 10 тыс. за кв.м. может встречаться в отдаленных мелких населенных пунктах, а цена выше 200 тыс. за кв.м. может быть у квартир в элитных жилых комплексах в престижных районах.
* Сильнее всего цена квартиры зависит от общей площади (коэффициент корреляции 0,79) и имеет четко выраженную линейную зависимость. Но в диапазоне: "площадь более 110 кв.м. - цена более 15 млн." зависмость перестаёт быть устойчивой. В классе дорогого элитного жилья площадь уже не является определяющим фактором, влияющим на цену. Большую роль играют прочие факторы, привносящие добавочную ценность: местоположение, дизайнерский ремонт и мебель, окружение инфраструктурой.
* Цена зависит от числа комнат (коэффициент корреляции 0,5): с увеличением числа комнат, квартира будет дороже. Самый большой разброс цен от недорогих до очень дорогих среди трехкомнатных квартир, а средяя цена по региону на них 7,375 млн. руб.
* Неярко выраженая обратная зависимость цены от удаленности от центра: чем квартира ближе к центру, тем дороже. Коэффициент корреляции:-0,34. На расстоянии от центра до 17 км наблюдается большой разброс цен (от 3 до 30 млн.руб.), после - большая часть цен укладывается в диапазон до 8 млн.руб. Квартиры в населенных пунктах области, находящихся более чем в 45 км.руб. от центра города, как правило относительно недорогие: до 6 млн.руб.
* Выражена зависимоть между ценой и этажом квартиры. Квартиры на первом этаже пользуются наименьшим спросом и как следствие дешевле прочих этажей на 24%. На первом этаже крайне мало дорогих квартир (более 10 млн.руб.). Чуть более привлекатен для покупки последний этаж. Он на 8% дешевле прочих этажей. 
* В выходные объявления публикуются реже, чем в будни. По воскресеньям объявленияй в 2,5 раза меньше, чем по четвергам.
* Зависимость средней цены от дня недели прослеживается. Квартиры, размещенные в объявлениях на выходных, продаются на 4-6% дешевле, чем те, что опубликованы в будни.
* Наблюдается зависимость цены от месяца размещения объявления: всплеск цен в апреле и сентябре и незначительное падение в июне. Колебания цен носят сезонный характер.
* Четко видна тенденция к росту популярности сервиса объявлений: рост более чем 2,5 раза в 2016 и 2017 годах. Однако, в 2018 году рынок интернет-объявлений насытился и его рост существенно замедлился.

* Наблюдается тенденция уменьшения доли очень дорогих квартир и уменьшения разницы в цене между квартирами в среднем по рынку.
* Из 10 населённых пунктов с наибольшим числом объявлений самая высокая цена за квадратный метр в Санкт_Петербурге - 110,3 тыс., а самая низкая в Выборге - 58,2 тыс.

Большинство квартир дороже 10 млн. так же расположены в центре. В то же время квартир стоимостью до 4,5 млн. в центре мало, а в городе такие квартиры максимально распространены.

В центре однокомнатные встречаются в 2 раза реже двух- и трехкомнатных и вполовину чаще четырехкомнатных. БОльшая часть 4-6ти комнатных квартир находится в центре города.

Причем в абсолютных цифрах разница увеличивается с ростом числа комнат до 3х, затем снижается. В среднем 3х комнатные квартиры в центре дороже, чем в городе на 2,66 млн., а однушки на 1,66 млн.

В городе зависимость имеет похожую картину, с незначительнвми отличиями. Квартиры на первом этаже в среднем дешевле везде, но в центре разница в цене выше, а в городе последние этажи ценятся немного выше.

В то же время квартир стоимостью до 4,5 млн. в центре мало, а в городе такие квартиры максимально распространены.

### Рекомендации

* Большое количество выбросов в данных о количестве дней публикации говорит о налии проблем при их учете. Отсутствие мониторинга причин, по которым объявления снимаются с публикации, может привести к серьезному искажению данных о времени продажи. Использование данных об объявлениях, снятых с публикации по иным причинам, кроме продажи, могли повлиять на результаты исследования. Следует обратить внимание на причины появления выбросов в этих данных, настроить алгоритмы БД на фильтрацию данных по причине снятия объявления с публикации и настроть выгрузку данны с учетом этой информации.
* Для улучшения качества предоставляемых для анализа данных рекомендуем улучшить контроль за полнотой и правильностью ввода данных в объявление продавцом. 
* Следует указать обязательными к заполнению такие поля, как: число балконов, число этажей, число комнат, названия населенных пунктов (если они вводятся в ручную).
* При вводе следует использовать стандартизированные списки названий населенных пунктов.
* Следует поставить фильтр на ввод нереалистичных данных с учетом выводов из проведенного исследования. К таким фильтрам следует отнестись с осторожностью, не перегружать обилием условий и слишком жесткими рамками, чтобы не допустить ухудшения пользовательского опыта и оттока клиентов. Например, количество комнат не должно быть меньше 1.
* Если для дальнейшего использования важна информация об удаленности от тех или иных объектов, следует проработать алгоритмы системы автоматизированного сбора данных локализации.
* Следует изучить причины возникновения пропусков данных о числе дней публикации. Если часть из них относятся к объявлениям, которые не сняты с публикации на момент выгрузки, необходимо наладить процесс выгрузки таким образом, чтобы избежать появления подобных проблем.
* При разработке автоматизированной системы для определения рыночной стоимости объекта недвижимости следует учитывать такие параметры как: общая площадь, число комнат, этаж, удаленность от центра. Эти факторы наибольшим образом влияют на цены. Следует учесть, что зависмость цен от удаленности от центра в самом центре не играет значительной роли.
* Следует также принять во внимание зависимость цены от дня недели публикации объявления. Для выходных цены в среднем на 4-6% ниже. Однако не стоит включать этот параметр для разработки системы, либо его удельный вес должен быть крайне незначительным.  
* При разработке автоматизированной системы не следует принимать во внимание сезонный фактор изменения цен. Его влияние не стабильно и непродолжительно во времени.
* Следует расчитывать рыночную цену исходя из данных не ранее последних 2-3 лет.
* Стоит дополнительно исследовать другие параметры, которые могут оказывать влияние на цены: расположение вблизи аэропорта, парков и водоемов. Или другие. Для более точного исследования необходимо уточнить пропущенные данные.
